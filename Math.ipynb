```python
import json

file_path = r"D:\Python\files\newfile.txt"   # change if needed

book = {} # dictionary
book['tom'] = {
    'name': 'tom',
    'address': ' 1 red street , NY',
    'phone': 8943989895
}
book['bob'] = {
    'name': 'bob',
   'address': ' 1 green street , NY',
    'phone': 89439487982
}

import json
s = json.dumps(book)
print(s)

# "w" mode ‚Üí clears old content and writes new
with open(file_path, "w") as f:
    json.dump(s, f, indent=4) # indent 4 is for common spacing f denotes format

print("JSON written successfully.")

```

```python
# file commands
'''
 | Mode | Meaning       | File must exist? | Overwrites?         | Pointer position | Read | Write        |
| ---- | ------------- | ---------------- | ------------------- | ---------------- | ---- | ------------ |
| `r`  | Read only     | ‚úÖ Yes            | ‚ùå No                | Start            | ‚úî    | ‚ùå            |
| `w`  | Write only    | ‚ùå No (creates)   | ‚úî Yes (clears file) | Start            | ‚ùå    | ‚úî            |
| `a`  | Append        | ‚ùå No (creates)   | ‚ùå No                | End              | ‚ùå    | ‚úî (add only) |
| `r+` | Read + Write  | ‚úÖ Yes            | ‚ùå No                | Start            | ‚úî    | ‚úî            |
| `w+` | Read + Write  | ‚ùå No (creates)   | ‚úî Yes (clears file) | Start            | ‚úî    | ‚úî            |
| `a+` | Read + Append | ‚ùå No (creates)   | ‚ùå No                | End              | ‚úî    | ‚úî (append)   |



'''
```

```python
# Importance of using with statement
'''
with is used to manage resources automatically (like files, database connections, locks, etc.).

üëâ Main purpose: Auto-close the file safely, even if an error occurs.

SCENARIO 1 :
f = open("file.txt", "r")
data = f.read()
f.close()   # You must remember this
=====================================================
SCENARIO 2 :
with open("file.txt", "r") as f:
    data = f.read()
'''
```
